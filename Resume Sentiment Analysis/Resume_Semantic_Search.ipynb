```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
resumes = [
    "Python developer experienced in machine learning, pandas and scikit-learn",

    "Frontend developer skilled in React, JavaScript and CSS",

    "Data analyst experienced in SQL, Excel, Power BI and Python",

    "Java backend developer experienced in Spring Boot and MySQL",

    "Machine learning engineer skilled in Python, TensorFlow and deep learning"
]
```

```python
resume_embeddings = model.encode(resumes)
```

```python
query = "Looking for a Python developer with machine learning experience"

query_embedding = model.encode([query])
```

```python
similarities = cosine_similarity(
    query_embedding,
    resume_embeddings
)[0]
```

```python
results = sorted(
    zip(resumes, similarities),
    key=lambda x: x[1],
    reverse=True
)

for resume, score in results:
    print(f"Score: {score:.4f}")
    print(resume)
    print()
```

```python
top_3 = results[:3]

for rank, (resume, score) in enumerate(top_3, start=1):
    print(f"Rank {rank}")
    print(f"Similarity: {score:.4f}")
    print(f"Resume: {resume}")
    print("-" * 50)
```
