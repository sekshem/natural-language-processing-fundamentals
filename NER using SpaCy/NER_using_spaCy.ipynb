```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
text = """
Saksham is a software engineer who studied at SRM Institute of Science
and Technology in Chennai. He has experience working with Python,
machine learning and data analytics.
"""

doc = nlp(text)
```

```python
for entity in doc.ents:
    print(
        "Entity:", entity.text,
        "| Label:", entity.label_,
        "| Meaning:", spacy.explain(entity.label_)
    )
```

```python
from spacy import displacy

displacy.render(
    doc,
    style="ent",
    jupyter=True
)
```
