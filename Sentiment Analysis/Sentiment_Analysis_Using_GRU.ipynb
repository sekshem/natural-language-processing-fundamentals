```python
import tensorflow as tf

from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense

print("TensorFlow version:", tf.__version__)
```

```python
# Number of most frequent words to keep
vocab_size = 10000

# Maximum length of each movie review
max_length = 200
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(
    num_words=vocab_size
)

print("Number of training samples:", len(x_train))
print("Number of testing samples:", len(x_test))

print("\nFirst review (integer tokens):")
print(x_train[0])

print("\nFirst review label:", y_train[0])
```

```python
x_train = pad_sequences(
    x_train,
    maxlen=max_length
)

x_test = pad_sequences(
    x_test,
    maxlen=max_length
)

print("Training data shape:", x_train.shape)
print("Testing data shape:", x_test.shape)
```

```python
model_gru = Sequential([
    Embedding(
        input_dim=vocab_size,
        output_dim=128
    ),

    GRU(
        64
    ),

    Dense(
        1,
        activation="sigmoid"
    )
])
```

```python
model_gru.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
model_gru.summary()
```

```python
history_gru = model_gru.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_split=0.2
)
```

```python
loss, accuracy = model_gru.evaluate(
    x_test,
    y_test
)

print("GRU Test Loss:", loss)
print("GRU Test Accuracy:", accuracy)
```

```python
predictions = model_gru.predict(x_test[:5])

for i, prediction in enumerate(predictions):
    probability = prediction[0]

    if probability >= 0.5:
        sentiment = "Positive"
    else:
        sentiment = "Negative"

    print(
        f"Review {i + 1}: "
        f"{sentiment} "
        f"(Probability: {probability:.4f})"
    )
```
