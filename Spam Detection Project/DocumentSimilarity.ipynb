```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
documents = [
    "Machine learning is transforming artificial intelligence.",
    "Artificial intelligence is changing the world using machine learning.",
    "Python is a popular programming language for data science.",
    "Deep learning is a subset of machine learning.",
    "The football match was exciting and full of energy."
]
```

```python
print("Documents:\n")

for i, doc in enumerate(documents, start=1):
    print(f"Document {i}:")
    print(doc)
    print()
```

```python
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
print("Vocabulary:\n")

print(vectorizer.get_feature_names_out())
```

```python
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out()
)

print(tfidf_df.round(2))
```

```python
similarity_matrix = cosine_similarity(tfidf_matrix)
```

```python
similarity_df = pd.DataFrame(
    similarity_matrix,
    columns=[f"Doc{i+1}" for i in range(len(documents))],
    index=[f"Doc{i+1}" for i in range(len(documents))]
)

print("\nCosine Similarity Matrix:\n")

print(similarity_df.round(2))
```

```python
highest_similarity = 0
doc_pair = ()

for i in range(len(documents)):
    for j in range(i + 1, len(documents)):

        if similarity_matrix[i][j] > highest_similarity:
            highest_similarity = similarity_matrix[i][j]
            doc_pair = (i + 1, j + 1)

print("\nMost Similar Documents")

print("-----------------------")

print(f"Document {doc_pair[0]} and Document {doc_pair[1]}")

print(f"Similarity Score: {highest_similarity:.2f}")
```

```python
doc1 = input("\nEnter Document 1:\n")

doc2 = input("\nEnter Document 2:\n")

new_documents = [doc1, doc2]

new_vectors = vectorizer.fit_transform(new_documents)

score = cosine_similarity(new_vectors)

print("\nSimilarity Score")

print(score[0][1])
```
